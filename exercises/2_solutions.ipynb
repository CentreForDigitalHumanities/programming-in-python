## 1

```python
exercise_list = [1, 12.9, "stressed", "bar", True]

for element in exercise_list:
    if element == "stressed":
        word = element

reversed_word = ""
for letter in word:
    reversed_word = letter + reversed_word
print(word + " in reverse is: " + reversed_word)
```

## 2

```python
numbers = [1, 5, -12, 38987, 0, 24]

total = 0
length = 0
for num in numbers:
    total = total + num
    length = length + 1
average = total/length
print(average)

# Adding strings makes the program fail, you cannot add a string to a number!

# Bonus: using builtin functions
total = sum(numbers)
length = len(numbers)
print(total/length)
```

## 3

```python
list1 = [1, 2, 3, 5, 1]
list2 = [1, 2, 3, 5, "1"],
list3 = [1, 2, 3, 5, True]
list4 = [False, "True", "False", "False"]
list5 = ["banana", "bananas", "Banana"]

master_list = [list1, list2, list3, list4, list5]
for sub_list in master_list:
    if sub_list[0] == sub_list[-1]:
        print("yay")
    else:
        print("nay")

# You might not expect list 3 to be a "yay"
# However, in Python False and True are equal to 0 and 1
print(False == 0)
print(True == 1)

# list5 is a nay, because strings are case sensitive
# if you want to compare them case-insensitive, first make then all lowercase
print("banana".lower() == "Banana".lower())
```

## 4

```python
# Many optimizations are possible!
numbers = list(range(0, 101))

for num in numbers:
    if num % 3 == 0 and num % 5 == 0:
        print("FizzBuzz")
    elif num % 3 == 0:
        print("Fizz")
    elif num % 5 == 0:
        print("Buzz")
    else:
        print(num)
```
