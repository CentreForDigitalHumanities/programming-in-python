# Python basics cheatsheet
## Types
#### numbers
```python
0
2.0
-8
```
####strings (text)
```python
# string (text)
"this is a string"
'this is also a string'
```
#### booleans
```python
True
False
```










## Operators
#### arithmetic
```python
1 + 1
8 - 2
10 * 4
12 / 3
```
#### string operators
```python
# concatenation
"hello" + "world"
# repetition
"hello"*3
```
#### comparison
```python
1 == 1 # equals
"hello" != "goodbye" # not equals
2 > 1 # greater than
1 < 2 # smaller than
```

## Variables
#### assigning and reassigning
```python
a = 20
a = "hello"
b = 2 + 1
c = 3 + b
```

## Lists
```python
[]
[1, 2, 3]
["word", True, 8.2]
```
#### accessing and setting elements
```python
list[index]
list[index] = "new value"
```

#### adding elements
```python
list = list + [element]
list.append(element)
list = list + [element1, element2]
```

#### slicing (x to y)
```python
list[x:y]
```

#### slicing (x to end)
```python
list[x:]
```


## Loops
#### loop over a list
```python
for element in list:
    # execute some code
    # and some more code
```
#### loop over a string
```python
for letter in "string":
    # execute some code
```
#### list of lists
```python
nested_list = [[element1, element2], [element3, element4]]
sublist1 = nested_list[0]
sublist2 = nested_list[1]

for sublist in nested_list: 
    for element in sublist:
        # execute some code
```






## Conditionals
#### if-elif-else
```python
if condition1:
    # execute some code
elif condition2:
    # execute some other code
else:
    # execute code if no conditions are true
```
#### logical operators
```python
condition1 and condition2
condition1 or condition2
not condition
```
