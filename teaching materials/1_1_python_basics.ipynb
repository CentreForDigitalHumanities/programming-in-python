* Naming
  * Rules
    * contains only `a-z`, `A-Z`, `0-9`, `_`
    * does not start with a number

    * meaningful name: `age = 33`, not `a = 30`

## 1.5 - Lists
- Collection of values
  - `[1, 2, 3]`, `["hello", "world"]`, `[]` (empty list)
  - Mixed types possible: `["hello", 1, False]`

```python
age = 33
age_list = [12, 20, age]
print(age_list)
```

```python
age = 33
age_someone_else = 34
list_of_ages = [age, age_someone_else, True, "hello", 2.0]
print(list_of_ages)
```

- Accessing list elements
  - *index*: from `0` to `length of list - 1`
  - Acces by index: `list[0]`, `list[1]`, `list[-1]` (last element)

```python
word_list = ["hello", "world"]
print(word_list[0])
world = word_list[1]
print(world)

word_list[-1]
```

  - Assign like any variable

```python
word_list = ["hello", "world"]
word_list[1] = "class"
print(word_list)
```

  - Add elements: `list + list` or `list.append(element)`


```python
list1 = ["hello", "world"]
list1 = list1 + ["!"]
print(list1)
```

```python
list1 = ['hello', 'world']
list2 = [1, 2]

big_list = list1 + list2
print(big_list)

big_list + ["another word"]
```

```python
list2 = [1, 2]
list2.append(3)
print(list2)
```

```python
list2.append(["hello", "world"])
print(list2)
print(list2[-1])
```

```python
list3 = [True, False]
list3 = list3 + False
print(list3)
```

  - Accessing multiple values ("slicing")
    - `list[begin:end]`
    - *end* is **not** included
    - `list[begin:]` everything from begin to the end of the list (end is included)

```python
list1 = ["hello", "world", "!"]
hello_world = list1[0:2]
print(hello_world)

list1[0:2] = ["hello", "class"]
print(list1)
```

```python
list1 = ["hello", "world", "!", "hello", "world", "!", "hello", "world", "!", "hello", "world", "!"]
```

## 1.6 - Loops
- Go through a list, run some code for each element

```python
list1 = ["hello", "world", "!"]


for value in list1:
    print(value)
```

```python
for word in list1:
    print(word)
```

- Multiple lines of code go in an indented block

```python
list1 = ["hello", "world", "!"]

for word in list1:
    new_word = word + "_new"
    print("printing the new word...")
    print(new_word)
    print(new_word)
```

- Loop over string


```python
for letter in "Hello, world!":
    print(letter)
```

- List of lists


```python
list_of_lists = [["hello", "world"], ["this", "is", "jelte"]]

for sublist in list_of_lists:
    print(sublist)
    for word in sublist:
        print(word)
```

```python
list_of_lists = [["hello", "world"], ["this", "is", "jelte"]]


for sublist in list_of_lists:
    print(sublist)
    for word in sublist:
        print(word)
```

## 1.7 - Conditionals
- Check if some statement is the case
- Run code if `True`

```python
a = 4

if a == 2:
    print("its 2")

if a == 4:
    print("a is four")

if a != 2:
    print("this will not get printed")

if a > 2:
    print("a is greater than 2")
```

- Add an `else` clause

```python
a = 4

if a == 2:
    print("a is two")
else:
    print("a is not two")
```

- Combining with lists

```python
list1 = [1, 2, 3, 4, 5]

for element in list1:
    if element == 3:
        print("we have found 3!")
    else:
        print("sadly, not 3")
```

- `elif` (else-if) for multiple `if` statements with shared `else`



```python
x = 6

if x == 3:
    print("x is 3")
elif x == 2:
    print("x is 2")
elif x == 5:
    print('nope')
else:
    print("x is neither 2 nor 3")
```
